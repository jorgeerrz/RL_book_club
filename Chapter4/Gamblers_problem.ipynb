In this notebook I implement value iteration for the gambler's problem from Chapter 4: Dynamic Programming

-Jorge Ramirez-Ruiz

So we want to define a finite MDP $\{S,A,P,R\}$, where:
- The set of states $S = \{0,...,100\}$ where 0 and 100 are terminal states.
- Action set for state $s$ is the amount that can be gambled $A(s) = \{0,1,...,\min(s,100-s)\}$. This minimum function comes from the fact that the maximum that the gambler can win is 100, so it only makes sense to gambleas much money as you can win.
- Probability of transition $\begin{equation}
p(s'|s,a) =
 \begin{cases} 
      p_h & s' = s + a \\
      1 - p_h & s' = s - a\\
      0 & otherwise
   \end{cases} 
\end{equation}$
- Rewards are zero for all transitions except when getting 100: $r(s = 100, a = 0) = 1$.


```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def adm_actions(s):
    return np.array(range(np.minimum(s,100-s)))+1
```

```python
def reward(s,a):
    r = 0
    if s == 100:
        r = 1
    return r
```

```python
def value_iteration(θ,p_h = 0.4):
    Δ = θ
    #Initialize values randomly
    V = np.random.rand(101)
    #Or zero
    #V = np.zeros(101) 
    #But values of terminal states are fixed to zero
    V[0] = V[100] = 0
    policy = np.zeros(101)
    #We can either say that the reward of getting to s = 100 is one
    #or reward for all transitions is zero, and value of state s = 100 is fixed to one
    #V[100] = 1
    while Δ >= θ:
        Δ = 0
        for s in range(1,100,1):
            v_temp = V[s]
            actions = adm_actions(s)
            q = np.zeros(np.size(actions))
            for idx,a in enumerate(actions):
                #For each action, only two states with nonzero probability
                q[idx] = p_h*(reward(s+a,a) + V[s+a]) + (1-p_h)*(reward(s-a,a) + V[s-a])
            #Use this code to break ties randomly between the best actions
            #maxs = np.flatnonzero(q == q.max())
            #opt_a = np.random.choice(maxs)
            opt_a = np.argmax(q)
            V[s] = q[opt_a]
            policy[s] = actions[opt_a]
            Δ = np.maximum(Δ, np.abs(V[s]-v_temp))
    return policy, V
```

```python
pol,V = value_iteration(1E-10,0.4)
```
